```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt

from google.colab import files

uploaded = files.upload()


data = pd.read_csv('LINTHALL.txt', delim_whitespace=True)

# Display the first few rows
print("Data head:")
print(data.head())

# Drop the columns
data_model = data.drop(['Obs', 'Loc', 'Type'], axis=1)

Y = data_model['BIO']
X = data_model.drop('BIO', axis=1)

X = sm.add_constant(X)

# Fit the OLS model
model = sm.OLS(Y, X).fit()

# Print full summary
print("\nOLS Regression Results:")
print(model.summary())

# Extract regression coefficients and standard errors
coefficients = model.params
std_errors = model.bse
SSE = np.sum(model.resid**2)

print("\nRegression Coefficients:")
print(coefficients)
print("\nStandard Errors:")
print(std_errors)
print("\nSum of Squared Errors (SSE): {:.4f}".format(SSE))

# Compute VIF for each predictor (excluding the intercept)
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                   for i in range(X.shape[1])]
print("\nVariance Inflation Factors (VIF):")
print(vif_data)

# Condition Number
X_no_const = X.drop('const', axis=1)
cond_number = np.linalg.cond(X_no_const)
print("\nCondition Number of X (without intercept): {:.4f}".format(cond_number))

# Eigenvalue decomposition
XtX = np.dot(X_no_const.T, X_no_const)
eigenvalues, _ = np.linalg.eig(XtX)
sorted_eigenvalues = np.sort(eigenvalues)
print("\nEigenvalues of X'X:")
print(sorted_eigenvalues)

if sorted_eigenvalues[0] < 1e-10:
    print("\nWarning: Near-zero eigenvalues detected, indicating severe collinearity.")
else:
    print("\nThe eigenvalues do not indicate severe collinearity.")
```

VIF Observations: several predictors exceed this threshold (for example, pH has a VIF of about 62, BUF about 34, Ca about 16.66, and Mg around 23.76).
These suggest that multicollinearity is present.

Condition Number Insight: The condition number is extremely high (approximately 66,205), which is well beyond the common threshold of 30.
This very large condition number reinforces the presence of multicollinearity.

Eigenvalue Analysis: the condition number derived from these eigenvalues (as noted above) is what really confuses me as they do not indicate collinearity.

```python

import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from google.colab import files


uploaded = files.upload()

data = pd.read_csv('LINTHALL.txt', sep='\s+')

# Display first few rows
print("Data head:")
print(data.head())

# Drop columns
data_model = data.drop(['Obs', 'Loc', 'Type'], axis=1)


Y = data_model['BIO']
X = data_model.drop('BIO', axis=1)


# Add constant
X_with_const = sm.add_constant(X)
ols_model = sm.OLS(Y, X_with_const).fit()

print("\n--- Full OLS Regression ---")
print(ols_model.summary())

coefficients_ols = ols_model.params
std_errors_ols = ols_model.bse
SSE_ols = np.sum(ols_model.resid**2)

print("\nOLS Regression Coefficients:")
print(coefficients_ols)
print("\nOLS Standard Errors:")
print(std_errors_ols)
print("\nOLS Sum of Squared Errors (SSE): {:.4f}".format(SSE_ols))
print("\nSum of OLS standard errors: {:.4f}".format(std_errors_ols.sum()))

# Standardize predictors
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
mu = scaler.mean_      # means of predictors
sigma = scaler.scale_  # std devs of predictors

# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# variance ratio and cumulative variance
explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)
print("\nExplained Variance Ratio by PC:")
print(np.round(explained_var, 4))
print("\nCumulative Explained Variance:")
print(np.round(cumulative_var, 4))

k = np.argmax(cumulative_var >= 0.95) + 1
print(f"\nNumber of principal components chosen (>=95% variance explained): {k}")

# Use only the first k principal components
X_pca_k = X_pca[:, :k]

# Add constant to the PC scores
X_pca_k_const = sm.add_constant(X_pca_k)

# Fit regression model
pcr_model = sm.OLS(Y, X_pca_k_const).fit()
print("\n--- PCR Regression Results (using first {} PCs) ---".format(k))
print(pcr_model.summary())

# Get PCR coefficients
gamma = pcr_model.params[1:]  # coefficients for PCs
gamma0 = pcr_model.params[0]  # intercept

# covariance matrix
cov_gamma = pcr_model.cov_params().iloc[1:, 1:]

V_k = pca.components_[:k].T

# Convert PCR coefficients back to original
beta_scaled = np.dot(V_k, gamma)

beta_original = beta_scaled / sigma
intercept_original = gamma0 - np.sum((mu / sigma) * beta_scaled)

coefficients_pcr = np.concatenate(([intercept_original], beta_original))
coef_index = ['const'] + list(X.columns)
coefficients_pcr = pd.Series(coefficients_pcr, index=coef_index)

print("\nPCR Regression Coefficients (converted back to original predictor scale):")
print(coefficients_pcr)

cov_beta_scaled = np.dot(np.dot(V_k, cov_gamma), V_k.T)
# Standard errors for original coefficients
se_beta = np.sqrt(np.diag(cov_beta_scaled)) / sigma
# standard error for intercept:
se_intercept = pcr_model.bse[0]
se_coefficients_pcr = np.concatenate(([se_intercept], se_beta))
se_coefficients_pcr = pd.Series(se_coefficients_pcr, index=coef_index)

print("\nPCR Standard Errors for Regression Coefficients (converted back to original scale):")
print(se_coefficients_pcr)

# Sum of Squared Errors (SSE)
Y_pred_pcr = intercept_original + np.dot(X, beta_original)
SSE_pcr = np.sum((Y - Y_pred_pcr)**2)
print("\nPCR Sum of Squared Errors (SSE): {:.4f}".format(SSE_pcr))

# Sum of standard errors
sum_se_pcr = se_coefficients_pcr.sum()
print("Sum of PCR standard errors: {:.4f}".format(sum_se_pcr))

print("\n--- Comparison ---")
print("OLS Sum of Standard Errors: {:.4f}".format(std_errors_ols.sum()))
print("PCR Sum of Standard Errors: {:.4f}".format(sum_se_pcr))
print("\nOLS SSE: {:.4f}".format(SSE_ols))
print("PCR SSE: {:.4f}".format(SSE_pcr))
```

In the OLS model, the sum of the standard errors of the estimated coefficients is very high (≈ 4048.09), and the sum of squared errors (SSE) is about 3.69×10⁶. In contrast, after applying Principal Components Regression (PCR) with collinearity reduction, the transformed coefficients in the original predictor space exhibit a dramatically lower sum of standard errors (≈ 174.70). However, the SSE for PCR increases to approximately 5.11×10⁶.

The reduction in the sum of standard errors indicates that PCR has substantially reduced the variance of the coefficient estimates and removed the multicollinearity. Overall, the PCR approach improved the stability and precision of the coefficient estimates by reducing multicollinearity (as shown with the lower standard errors), but this improvement comes at the cost of an increased SSE, indicating a loss in predictive accuracy.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
from itertools import combinations
from sklearn.linear_model import Ridge, RidgeCV
from google.colab import files

uploaded = files.upload()

data5 = pd.read_csv('LINTH-5.txt', sep='\s+')
print("Data head:")
print(data5.head())

data5 = data5[['BIO','SAL','pH','K','Na','Zn']]

X5 = data5[['SAL','pH','K','Na','Zn']]
X5_const = sm.add_constant(X5)

def collinearity_diagnostics(X):
    # Compute VIF
    vif_df = pd.DataFrame()
    vif_df["Variable"] = X.columns
    vif_df["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    X_no_const = X.drop('const', axis=1) if 'const' in X.columns else X
    cond_number = np.linalg.cond(X_no_const)

    # Eigenvalue decomposition
    XtX = np.dot(X_no_const.T, X_no_const)
    eigenvalues = np.sort(np.linalg.eigvals(XtX))

    print("\n--- Collinearity Diagnostics ---")
    print("VIFs:")
    print(vif_df)
    print(f"\nCondition Number (without intercept): {cond_number:.4f}")
    print("Eigenvalues of X'X:")
    print(eigenvalues)

collinearity_diagnostics(X5_const)

def stepwise_selection(X, y,
                       initial_list=[],
                       threshold_in=0.12,
                       threshold_out=0.12,
                       verbose=True):
    included = list(initial_list)
    while True:
        changed = False
        # forward step
        excluded = list(set(X.columns) - set(included))
        new_pval = pd.Series(index=excluded, dtype=float)
        for new_column in excluded:
            model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included + [new_column]]))).fit()
            new_pval[new_column] = model.pvalues[new_column]
        if not new_pval.empty:
            best_pval = new_pval.min()
            if best_pval < threshold_in:
                best_feature = new_pval.idxmin()
                included.append(best_feature)
                changed = True
                if verbose:
                    print(f"Add {best_feature:>5} with p-value {best_pval:.4f}")

        # backward step
        model = sm.OLS(y, sm.add_constant(pd.DataFrame(X[included]))).fit()
        # Exclude intercept\
        pvalues = model.pvalues.iloc[1:]
        worst_pval = pvalues.max()
        if worst_pval > threshold_out:
            worst_feature = pvalues.idxmax()
            included.remove(worst_feature)
            changed = True
            if verbose:
                print(f"Remove {worst_feature:>5} with p-value {worst_pval:.4f}")
        if not changed:
            break
    return included

print("\n--- Stepwise Regression ---")
X5_df = X5.copy()
y5 = data5['BIO']
selected_features = stepwise_selection(X5_df, y5, threshold_in=0.12, threshold_out=0.12)
print("\nFinal selected features:", selected_features)

# Fit final model
X5_selected = sm.add_constant(X5_df[selected_features])
stepwise_model = sm.OLS(y5, X5_selected).fit()
print("\nFinal Stepwise Regression Model Summary:")
print(stepwise_model.summary())

# collinearity diagnostics
collinearity_diagnostics(X5_selected)


print("\n--- Ridge Regression ---")

alphas = np.logspace(-2, 4, 100)
coefs = []

X5_array = X5.values
for a in alphas:
    ridge = Ridge(alpha=a, fit_intercept=True)
    ridge.fit(X5_array, y5)
    coefs.append(ridge.coef_)

coefs = np.array(coefs)

# Plot ridge trace
plt.figure(figsize=(8,6))
for i, col in enumerate(X5.columns):
    plt.plot(alphas, coefs[:, i], label=col)
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('Coefficient')
plt.title('Ridge Trace')
plt.legend()
plt.show()

ridge_cv = RidgeCV(alphas=alphas, store_cv_values=True)
ridge_cv.fit(X5_array, y5)
print(f"Selected alpha (RidgeCV): {ridge_cv.alpha_:.4f}")
print("Ridge Coefficients:", dict(zip(X5.columns, ridge_cv.coef_)))

# refit the model with the full set.
X5_ridge = sm.add_constant(X5)
ridge_model = sm.OLS(y5, X5_ridge).fit()  # using OLS on selected predictors as a check
print("\nRefitted Model after Ridge-based Variable Selection (if any variables were dropped):")
print(ridge_model.summary())
collinearity_diagnostics(X5_ridge)


print("\n--- Subset Selection (Best Two-Variable Model) ---")
predictors = list(X5.columns)
results = []
for combo in combinations(predictors, 2):
    X_subset = sm.add_constant(X5[list(combo)])
    model = sm.OLS(y5, X_subset).fit()
    aic = model.aic
    bic = model.bic
    sse = np.sum(model.resid**2)
    # sum of VIF
    vif_sum = np.sum([variance_inflation_factor(X_subset.values, i)
                      for i in range(1, X_subset.shape[1])])  # exclude constant
    results.append({
        'Predictors': combo,
        'AIC': aic,
        'BIC': bic,
        'SSE': sse,
        'VIF_sum': vif_sum
    })
results_df = pd.DataFrame(results)
print(results_df.sort_values(by='AIC'))

print("\nBest model by AIC:")
best_aic = results_df.loc[results_df['AIC'].idxmin()]
print(best_aic)

print("\nBest model by BIC:")
best_bic = results_df.loc[results_df['BIC'].idxmin()]
print(best_bic)

print("\nBest model by SSE:")
best_sse = results_df.loc[results_df['SSE'].idxmin()]
print(best_sse)
```

both the stepwise regression and the subset selection methods independently identified the (pH, Na) model as the best two-variable model. Therefore, we can concluse these two predictors are the most stable and relevant for predicting BIO in the reduced data set.

It's interesting to note that the full five-predictor model had a very large condition number and higher VIFs, indicating that multicollinearity was a significant problem. The ridge regression analysis shows that with this full five predictors, an optimal penalty would be α ≈ 2.6561. However, the model with pH and Na balances model simplicity with predictive performance while minimizing collinearity. This is shown also with the agreement between AIC, BIC, and SSE.
